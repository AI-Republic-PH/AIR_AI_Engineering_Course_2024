```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load the T5 model and tokenizer (small version)
model_name = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Set up a few-shot prompt for English-to-Tagalog translation
prompt = """
Translate the following English sentences to Tagalog:
1. How are you? -> Kamusta ka?
2. I love to eat mangoes. -> Mahilig akong kumain ng mangga.
3. The sun is shining brightly today. -> Maliwanag ang sikat ng araw ngayon.
4. We are going to the market. ->
"""

# Tokenize the input
inputs = tokenizer(prompt, return_tensors="pt")

# Generate the output using the model
outputs = model.generate(
    inputs['input_ids'],
    max_length=50,          # Maximum length of the generated output
    num_beams=5,            # Beam search for better generation
    early_stopping=True     # Stop when the end of the sequence is predicted
)

# Decode and print the output
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated Text:")
print(generated_text)
```
